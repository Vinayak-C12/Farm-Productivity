```python
def change_units(df):
    if df['Production Units']=='Bales':
        data=df['Production']*1.7
        return data
    elif df['Production Units'] == 'Nuts':
        data = df['Production'] / 50
        return data
    else:
        return df['Production']
```

```python
df["Production"] = df.apply(change_units,axis=1)
```

```python
# crop type with maximum production over the years (considering production in tonnes) - Top 10

def group_and_order(data, group_by_column, order_by_column, ascending=True):
    grouped_data = data.groupby(group_by_column)[order_by_column].max()
    sorted_data = grouped_data.sort_values(ascending=ascending)
    return sorted_data

max_crop_yield = group_and_order(data, 'Crop', 'Production', ascending=False)

max_crop_yield.head(10)
```

```python
# Assuming you have a DataFrame named 'data' with columns 'Year', 'Crop', 'Production', and 'Yield'

# Extract data for Coconut production
coconut_data = data[data['Crop'] == 'Coconut']

# Group by Year and calculate the mean production
coconut_prod_by_year = coconut_data.groupby('Year')['Production'].mean()

# Plotting the graph for Coconut production
plt.plot(coconut_prod_by_year.index.to_numpy(), coconut_prod_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Mean Crop Production of Coconut Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Group by Year and calculate the mean yield
coconut_yield_by_year = coconut_data.groupby('Year')['Yield'].mean()

# Plotting the graph for Coconut yield
plt.plot(coconut_yield_by_year.index.to_numpy(), coconut_yield_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Yield in Tonnes')
plt.title('Mean Crop Yield of Coconut Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

coconut_yield_by_state = coconut_data.groupby('State')['Yield'].mean()

# Plotting the graph for Coconut yield by state
plt.figure(figsize=(12, 8))
sns.barplot(x=coconut_yield_by_state.index, y=coconut_yield_by_state.values)
plt.xlabel('State')
plt.ylabel('Mean Yield in Tonnes')
plt.title('Mean Crop Yield of Coconut by State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Assuming you have a DataFrame named 'data' with columns 'Year', 'Crop', 'Production', and 'Yield'

# Extract data for Sugarcane production
sugarcane_data = data[data['Crop'] == 'Sugarcane']

# Group by Year and calculate the mean production
sugarcane_prod_by_year = sugarcane_data.groupby('Year')['Production'].mean()

# Plotting the graph for Sugarcane production
plt.plot(sugarcane_prod_by_year.index.to_numpy(), sugarcane_prod_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Mean Crop Production of Sugarcane Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Group by Year and calculate the mean yield
sugarcane_yield_by_year = sugarcane_data.groupby('Year')['Yield'].mean()

# Plotting the graph for Sugarcane yield
plt.plot(sugarcane_yield_by_year.index.to_numpy(), sugarcane_yield_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Yield in Tonnes')
plt.title('Mean Crop Yield of Sugarcane Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


sugarcane_yield_by_state = sugarcane_data.groupby('State')['Yield'].mean()

# Plotting the graph for Coconut yield by state
plt.figure(figsize=(12, 8))
sns.barplot(x=sugarcane_yield_by_state.index, y=sugarcane_yield_by_state.values)
plt.xlabel('State')
plt.ylabel('Mean Yield in Tonnes')
plt.title('Mean Crop Yield of Sugarcane by State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Let's build a matrix to see over the years what is the mean production in tonnes of 
# all the crops in all the states

production_matrix = data.pivot_table(index = 'Crop', columns = 'State', values = 'Production', aggfunc = np.mean).round(0).fillna(0)

print("Mean Crop Production (in Tonnes) in States - Matrix :")
print("\n")
production_matrix
```

```python
# A Crop Yield x Season Matrix can help us understand what's the best year for production of any crop.
season_matrix = data.pivot_table(index = 'Crop', columns = 'Season', values = 'Production', aggfunc = np.mean).round(0).fillna(0)
season_matrix['Grand_Total'] = season_matrix.sum(axis = 1)

season_matrix = season_matrix.sort_values('Grand_Total', ascending = False ) 

print("Mean Crop Production (in Tonnes) in Seasons - Matrix :")
print("\n")
season_matrix
```

```python
# Let's see which state has the max production in Tonnes in a given year

year = '2011-12'

def get_state_production(year):
    filtered_data = data[data['Year'].str.contains(year)]
    state_production = filtered_data.groupby('State')['Production'].sum().astype(int)
    state_production_sorted = state_production.sort_values(ascending=False)
    return state_production_sorted

sorted_state_production = get_state_production(year)

sorted_state_production.head(3)
```

```python
# Next we will build a Pi-chart of top 5 crops in terms of 
# production volume in tonnes of user enetered state and year

class StateProductionChart:
    
    def create_pie_chart(state, year):
        filtered_data = data[(data['State'] == state) & (data['Year'] == year)]
        group_data = filtered_data.groupby('Crop')['Production'].sum()
        top_5_crops = group_data.nlargest(5) #change number here to get more crops in pie-chart
        
        sizes = list(top_5_crops.values)
        labels = top_5_crops.index
        
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        plt.axis('equal')
        plt.title(f'Crop Production in {state} (Year: {year})')
        plt.show()

#copy paste the below 3 lines to get multiple pie-charts

# Pie-Chart One
state_name = 'Karnataka'
year = '2010-11'

StateProductionChart.create_pie_chart(state_name,year)

# Pie-Chart Two
state_name = 'Uttar Pradesh'
year = '2015-16'

StateProductionChart.create_pie_chart(state_name,year)
```
